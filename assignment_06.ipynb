```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("weather.csv")
```

```python
data=data.dropna()
```

```python
sns.distplot(data['Evaporation'])
```

```python
sns.distplot(data['Evaporation'],kde=False)
```

```python
sns.distplot(data['Evaporation'],kde=False,rug=True)
```

```python
sns.distplot(data['Rainfall'],hist=False,rug=True)
```

```python
sns.kdeplot(data['MaxTemp'],shade=True)
```

```python
sns.jointplot(x=data['MaxTemp'],y=data['MinTemp'])
```

```python
with sns.axes_style('white'):
    sns.jointplot(x = data['Temp3pm'], y = data['Temp9am'], kind = 'hex', color = 'k')
```

```python
sns.jointplot(x='Cloud9am', y='Temp9am', data=data, kind='kde')
```

```python
plt.scatter(x=data['Temp9am'],y=data['Temp3pm'])
```

```python
df=pd.DataFrame(data,columns=['MaxTemp','Evaporation','Sunshine','Rainfall'])
sns.pairplot(df)
```

```python
sns.stripplot(x ='Cloud3pm' , y='Humidity3pm', data=data, jitter=True)
```

```python
sns.stripplot(x ='Cloud3pm' , y='Humidity3pm', data=data)
```

```python
sns.boxplot(x ='Cloud9am' , y='Humidity9am', data=data)
```

```python
sns.boxplot(x ='Cloud3pm' , y='Humidity3pm', data=data,hue='RainToday', dodge=True)
```

```python
sns.barplot(x='RainToday', y='Cloud3pm', data=data)
```

```python
sns.barplot(x='Cloud3pm', y='Humidity9am', data=data, hue='RainToday')
```

```python
sns.countplot(x='RainToday', data=data, palette='Greens_d')
```

```python
sns.swarmplot(x='Cloud9am', y='Humidity9am', data=data)
```

```python
df=pd.DataFrame(data,columns=['MaxTemp','Evaporation','Sunshine','RainToday'])
sns.pairplot(df,hue='RainToday')
```

```python
sns.kdeplot(data['MaxTemp'], label='original')
sns.kdeplot(data['MaxTemp'], bw=0.8, label='bw=0.8')
sns.kdeplot(data['MaxTemp'], bw=5, label='bw=5')
plt.legend()
```
